```python
# What is the most common way the players got out in the IPL? How many?
import pandas as pd
import numpy as np

deliveries = pd.read_csv("deliveries.csv")

deliveries.groupby(["dismissal_kind"]).batsman.value_counts().sort_values().tail(1)
```

```python
# Add a column called 'Howzthat' that has shows 'Yay' if the corresponding player_dismissed column has non NaN value
# and 'Nay' if player_dismissed is NaN. Hint: lookat np.where()
import pandas as pd
import numpy as np
deliveries = pd.read_csv("deliveries.csv")
deliveries['Howzthat']=np.where(deliveries['player_dismissed'].isnull(),"Nay","Yay")
deliveries
```

```python
#10. Add a column 'run_run' which checks total_runs column and say 
#'Duck' if total_runs is 0, 
#'come_on' if total_runs is 1 or 2,
# 'cool' if total_runs is 3,
# 'hit!' if total_runs is 4 or 6. 
# Hint: you can write a function using nested if-else and apply that function to df.total_runs
import pandas as pd
import numpy as np

deliveries = pd.read_csv("deliveries.csv")
def comment(total_runs):
    if total_runs==0:
        return("Duck")
    elif(total_runs==1 or total_runs==2):
        return("come_on")
    elif total_runs==3:
        return("cool")
    elif(total_runs==4 or total_runs==6):
        return("hit!")
    
deliveries['run_run']=deliveries.total_runs.apply(comment)
deliveries
    

```
